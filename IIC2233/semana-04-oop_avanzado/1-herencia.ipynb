# Herencia

La **herencia** (*inheritance*) es una de las caracterÃ­sticas mÃĄs importantes de OOP, y corresponde a una relaciÃģn de **especializaciÃģn** y **generalizaciÃģn** entre clases. En esta relaciÃģn, una **clase** _hereda_ atributos y mÃĐtodos de otra. Decimos que la que hereda es una **subclase**, y la otra es una **superclase**. La subclase posee todos los atributos y mÃĐtodos de la superclase, pero ademÃĄs tiene sus propios mÃĐtodos y atributos especÃ­ficos, que se consideran una especializaciÃģn. El concepto de herencia nos permite aprovechar (reutilizar) cÃģdigo de las clases de las cuales se hereda.

## Ejemplo: especializando la clase `Auto`

Consideremos una clase `Auto`. Algunos de sus atributos son `marca`, `modelo`, `color` y algunos de sus mÃĐtodos son `conducir`, `leer_odÃģmetro` y `vender`.

```python
class Auto:

    def __init__(self, marca, modelo, aÃąo, color, km):
        self.marca = marca
        self.modelo = modelo
        self.aÃąo = aÃąo
        self.color = color
        self._kilometraje = km
        self._dueÃąo = None

    def conducir(self, kms):
        print(f"Conduciendo {kms} kilÃģmetros")
        self._kilometraje += kms

    def vender(self, nuevo_dueÃąo):
        self._dueÃąo = nuevo_dueÃąo
        print(f"Auto vendido a {nuevo_dueÃąo}")

    def leer_odometro(self):
        return self._kilometraje
```

Si se nos pide modelar un furgÃģn escolar, es natural pensar que ÃĐste compartirÃĄ muchas caracterÃ­sticas de un `Auto`. Es mÃĄs, podrÃ­amos decir que un furgÃģn escolar **es un tipo particular de `Auto`**. El furgÃģn escolar deberÃ­a tener, al menos, _los mismos atributos y mÃĐtodos_ que un auto, y seguramente tendrÃĄ tambiÃĐn algunos atributos adicionales como _lista de estudiantes inscritos_ y mÃĐtodos adicionales como _inscribir estudiante_. Como ya hemos escrito el cÃģdigo de la clase `Auto`, nos gustarÃ­a poder **reutilizar** este cÃģdigo, pues ahÃ­ ya tenemos algunos atributos y mÃĐtodos implementados.

Usaremos, entonces, la **herencia**. La herencia nos permite _heredar_ datos y comportamientos de una clase y utilizarlos en otra. En nuestro ejemplo del furgÃģn escolar, crearemos una clase `FurgÃģnEscolar` que hereda de `Auto` y definiremos ahÃ­ la lista de estudiantes, y un mÃĐtodo de inscripciÃģn.

Si `FurgÃģnEscolar` **hereda** de `Auto`, tambiÃĐn se dice que:
- `FurgÃģnEscolar` es una **especializaciÃģn** de la clase `Auto`
- `FurgÃģnEscolar` es una **subclase** (o clase hija) de `Auto`
- `FurgÃģnEscolar` **extiende** a la clase `Auto`
- `Auto` es **superclase** (o clase madre) de `FurgonEscolar`

En Python, la implementaciÃģn de herencia se describe especificando, al momento de definir una clase, el nombre de la clase de la cual hereda.

```python
class FurgonEscolar(Auto): # AquÃ­ se marca de donde hereda
    """Subclase de Auto"""
    
    def __init__(self, marca, modelo, aÃąo, color, km):
        # Para inicializar algunos datos en la clase madre, llamamos explÃ­citamente 
        # al __init__ de esa clase. Lo llamaremos usando la clase madre
        Auto.__init__(self, marca, modelo, aÃąo, color, km)
        
        # Este atributo existe Ãšnicamente para objetos de tipo FurgonEscolar, 
        # pero no para todos los objetos de clase Auto 
        self.estudiantes = []
    
    # inscribir_estudiante es un mÃĐtodo especÃ­fico de esta subclase.
    def inscribir_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)
```

Notemos que el inicializador de `FurgonEscolar` no inicializa los atributos `marca`, `modelo`, `aÃąo`, `color` y `_kilometraje`. PodrÃ­a hacerlo, pero es mÃĄs conveniente que los atributos que ya son inicializados por la clase superior se le entreguen a su inicializador respectivo. Para invocar al inicializador de la clase superior debemos llamar explÃ­citamente el nombre de dicha clase antes del `__init__`; de lo contrario estarÃ­amos llamando recursivamente al mismo `__init__` de la clase actual. Una vez que se ha llamado al `__init__` de la clase superior, podemos continuar inicializando los atributos Ãšnicos de `FurgÃģnEscolar`.

Podemos comprobar que un objeto de la clase `FurgÃģnEscolar` tiene todos los datos y mÃĐtodos que tenÃ­a la clase `Auto`, y tambiÃĐn tiene lo que definimos especÃ­ficamente para `FurgonEscolar`:

```python
furgon = FurgonEscolar('Kia', 'Sportage', 2000, "Blanco", 135000)
print(f"Marca: {furgon.marca}")
print(f"Modelo: {furgon.modelo}")
print(f"Color: {furgon.color}")
furgon.conducir(5)
print(f"Kilometraje: {furgon.leer_odometro()}")
furgon.inscribir_estudiante('Jaimito')
furgon.inscribir_estudiante('Danielita')
furgon.inscribir_estudiante("ToÃąito")
furgon.inscribir_estudiante('Pablito')
furgon.inscribir_estudiante('Tamita')
print(f"Estudiantes: {furgon.estudiantes}")
```

## Sobreescritura de mÃĐtodos: *overriding*

En ocasiones, algunos mÃĐtodos de la superclase (que son heredados por la subclase) no nos sirven exactamente para la subclase y podrÃ­amos querer reemplazar o modificar algunos de ellos. En este ejemplo, no queremos que el conductor de un `FurgonEscolar` conduzca igual que un conductor de `Auto`, si no que queremos que lo haga con mÃĄs cuidado. DespuÃĐs de todo, lleva estudiantes.

La herencia nos permite **sobreescribir** los mÃĐtodos que necesitemos modificar. En Python, podemos **volver a definir un mÃĐtodo en una subclase**, con el mismo nombre que tenÃ­a en la superclase. De esta manera al ejecutar el mÃĐtodo desde un objeto perteneciente a la subclase, se ejecuta la versiÃģn del mÃĐtodo que estÃĄ en la subclase, y no la versiÃģn que estÃĄ en la sÃšperclase. En OOP, esto se llama _**overriding**_ y permite que una subclase implemente una versiÃģn especializada de un mÃĐtodo que es heredado desde una superclase.

Revisemos de nuevo el ejemplo del `Auto` y `FurgonEscolar`:


```python
class Auto:
    
    def __init__(self, ma, mo, a, c, k):
        self.marca = ma
        self.modelo = mo
        self.aÃąo = a
        self.color = c
        self._kilometraje = k
        self._dueÃąo = None

    def conducir(self, kms):
        print(f"Conduciendo {kms} kilÃģmetros")
        self._kilometraje += kms

    def vender(self, nuevo_dueÃąo):
        self._dueÃąo = nuevo_dueÃąo
        print(f"Auto vendido a {nuevo_dueÃąo}")

    def leer_odometro(self):
        return self._kilometraje

    
class FurgonEscolar(Auto):
    """Subclase de Auto"""
    
    # Estamos haciendo overriding del __init__ original
    def __init__(self, marca, modelo, aÃąo, color, kms):
        # AÃšn queremos usar el __init__ original para setear los otros datos. AsÃ­ es como podemos llamarlo.
        Auto.__init__(self, marca, modelo, aÃąo, color, kms)
        self.estudiantes = []
    
    # inscribir_estudiante es un mÃĐtodo especÃ­fico de esta subclase.
    def inscribir_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)
        
    # Estamos haciendo overriding del mÃĐtodo conducir original
    def conducir(self, distancia):
        # AcÃĄ no queremos usar la versiÃģn original de conducir
        print(f"Conduciendo con cuidado {distancia} kilÃģmetros")
```

En este caso hemos reescrito el mÃĐtodo `conducir`, que ya estaba definido en la clase `Auto`, pero ahora dentro de la clase `FurgonEscolar`. Hemos hecho un _override_ del mÃĐtodo `conducir`. De esta manera si llamamos a `conducir` desde una instancia de `FurgonEscolar`, se ejecutarÃĄ el `conducir` que acabamos de reescribir, pero si llamamos a `conducir` desde una instancia de `Auto`, se ejecutarÃĄ el `conducir` de `Auto`.

```python
auto = Auto('Suzuki', 'Grand Nomade', 2015, "Naranjo", 35000)
furgon = FurgonEscolar('Kia', 'Sportage', 2000, "Blanco", 135000)
print(f"Marca: {furgon.marca}")
print(f"Modelo: {furgon.modelo}")
print(f"Color: {furgon.color}")
furgon.conducir(5)
print(f"Kilometraje: {furgon.leer_odometro()}")
furgon.inscribir_estudiante("Jaimito")
furgon.inscribir_estudiante("Danielita")
furgon.inscribir_estudiante("ToÃąito")
furgon.inscribir_estudiante("Pablito")
furgon.inscribir_estudiante("Tamita")
print(f"Estudiantes: {furgon.estudiantes}")
auto.conducir(12)
print(f"Tipo de auto: {type(auto)}")
print(f"Tipo de furgÃģn: {type(furgon)}")
```

Podemos ver que, al ejecutar el mismo mÃĐtodo sobre `auto` y sobre `furgon`, el resultado es distinto. Al ejecutar `furgon.conducir(5)`, se ejecuta el mÃĐtodo _especializado_ que se definiÃģ para la clase `FurgonEscolar`, mientras que si ejecutamos `auto.conducir(12)`, se ejecuta el mÃĐtodo original de la clase `Auto`.

## Obtener clase superior: `super()`

En el ejemplo del furgÃģn escolar, reimplementamos (hicimos _override_ de) el mÃĐtodo inicializar `__init__`. Esto lo hacemos porque tambiÃĐn queremos inicializar el nuevo atributo `estudiantes`.  Pero tambiÃĐn debemos inicializar los atributos heredados de la clase `Auto`. Si no queremos hacer ninguna modificaciÃģn a la manera en que ÃĐstos se inicializan, podemos llamar explÃ­citamente al mÃĐtodo de la superclase. Si escribimos una lÃ­nea de la forma `SuperClase.metodo(self, argumentos)`, podemos invocar directamente la ejecuciÃģn del mÃĐtodo de la superclase. En nuestro caso, podemos ejecutar `Auto.__init__(self, marca, modelo, motor)` y con esto estamos delegando la inicializaciÃģn de esos atributos a la superclase.

Usando el mÃĐtodo `super()` podemos utilizar la implementaciÃģn de un mÃĐtodo de la superclase sin nombrar explÃ­citamente a la clase superior, de la forma `super().metodo(argumentos)`. Esto nos ayuda a escribir un cÃģdigo mÃĄs mantenible, en caso que decidamos cambiar el nombre de la clase superior. AdemÃĄs, evitarÃĄ problemas al momento de tener mÃšltiples herencias, donde muchas cosas interesantes pueden pasar. Un buen diseÃąo dicta que este mÃĐtodo tenga la misma estructura de llamada en todos los casos.

```python
class FurgonEscolar(Auto):
    """Subclase de Auto"""
    
    # Estamos haciendo overriding del __init__ original
    def __init__(self, marca, modelo, aÃąo, color, kms):
        # AÃšn queremos usar el __init__ original para setear los otros datos. AsÃ­ podemos llamarlo con super()
        super().__init__(marca, modelo, aÃąo, color, kms)
        self.estudiantes = []
    
    # inscribir_estudiante es un mÃĐtodo especÃ­fico de esta subclase.
    def inscribir_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)
        
    # Estamos haciendo overriding del mÃĐtodo conducir original
    def conducir(self, distancia):
        # AcÃĄ no queremos usar la versiÃģn original de conducir
        print(f"Conduciendo con cuidado {distancia} kilÃģmetros")  


furgon = FurgonEscolar('Kia', 'Sportage', 2000, "Blanco", 135000)
print(f"Marca: {furgon.marca}")
print(f"Modelo: {furgon.modelo}")
print(f"Color: {furgon.color}")
furgon.conducir(5)
print(f"Kilometraje: {furgon.leer_odometro()}")
furgon.inscribir_estudiante("Jaimito")
furgon.inscribir_estudiante("Danielita")
furgon.inscribir_estudiante("ToÃąito")
furgon.inscribir_estudiante("Pablito")
furgon.inscribir_estudiante("Tamita")
print(f"Estudiantes: {furgon.estudiantes}")
```

## Ejemplo: Herencia con _built-ins_

La herencia nos permite extender, no solamente las clases que hemos definido, sino tambiÃĐn algunas que ya existÃ­an. En Python le llamamos clases *built-ins* a aquellas que estÃĄn _construidas dentro_ ("*built*"-"*in*") del intÃĐrprete de Python y no necesitan de ningÃšn mÃģdulo adicional para funcionar. 

Una de las clases *built-in* de Python es la clase `list`. Si queremos extender la clase `list`, podemos definir una subclase que heredarÃĄ los mÃĐtodos de la clase `list` y a su vez tendrÃĄ datos y mÃĐtodos propios:

```python
class ListaDeContactos(list):
    """
    Estamos extendiendo y especializando la clase list estÃĄndar. 
    Tiene todos los mÃĐtodos de la lista mÃĄs los definidos por nosotros.
    """
    
    # Buscar es un mÃĐtodo especÃ­fico de esta sub-clase
    def buscar(self, nombre):
        matches = []
        for contacto in self:
            if nombre in contacto.nombre:
                matches.append(contacto)
        return matches

    
class Contacto:
    """La clase Contacto almacena nombre y correo electrÃģnico."""
    
    def __init__(self, nombre, email):
        self.nombre = nombre
        self.email = email


class Familiar(Contacto):
    """Familiar es una clase especializada de Contacto que permite incluir el tipo de relaciÃģn"""

    # Overriding sobre el mÃĐtodo __init__()
    def __init__(self, nombre, email, relacion):
        super().__init__(nombre, email)
        self.relacion = relacion
```

```python
contactos_lista = ListaDeContactos()
contactos_lista.append(Familiar(nombre="Catalina Kendrick", email="cata.k@gmail.com", relacion="prima"))
contactos_lista.append(Contacto(nombre="Ami Carol", email="ami.carol@gmail.com"))
contactos_lista.append(Familiar(nombre="Carol Fernandez", email="klaus@fernandez.cl", relacion="Hermana"))
contactos_lista.append(Contacto(nombre="Fernando Laftrache", email="fernando@laftrache.com"))

personas_llamadas_carol = []
for contacto in contactos_lista.buscar("Carol"):
       personas_llamadas_carol.append(contacto.nombre)
print(personas_llamadas_carol)
```

En este ejemplo, la clase `ListaDeContactos` extiende a `list` para agregar un mÃĐtodo que busca sobre sÃ­ misma (`self`) todos los elementos que coincidan con cierto _string_. Una vez creado un objeto de tipo `ListaDeContactos`, este objeto posee el mÃĐtodo `buscar`.
